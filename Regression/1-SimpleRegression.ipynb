# 1. Import Libraries and Generate Data

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

# Generate synthetic regression data
X, y = make_regression(n_samples=100, n_features=1, noise=20, random_state=0)
X = X.ravel()  # Flatten X to 1D array for easier manipulation
```

# 2. Calculate Simple Linear Regression Coefficients

```python
# Calculate means of X and y
x_bar, y_bar = X.mean(), y.mean()

# Calculate slope (beta) and intercept (alpha)
beta = ((X - x_bar) * (y - y_bar)).sum() / ((X - x_bar)**2).sum()
alpha = y_bar - beta * x_bar

# Generate predicted values using the regression equation
y_pred = alpha + beta * X
```

# 3. Visualize Data and Regression Line

```python
plt.figure(figsize=(6, 4))
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.title('Simple Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

# 4. Calculate Performance Metrics

```python
# Residual Sum of Squares (RSS)
RSS = ((y - y_pred) ** 2).sum()

# Total Sum of Squares (TSS)
TSS = ((y - y.mean()) ** 2).sum()

# Coefficient of Determination (RÂ²)
R_squared = 1 - (RSS / TSS)

# Mean Squared Error (MSE)
mse = ((y_pred - y) ** 2).mean()

# Root Mean Squared Error (RMSE)
rmse = mse ** 0.5

# Residuals (errors)
residuals = y - y_pred

# Print metrics with explanation
print(f"R-squared (explained variance): {R_squared:.2f}")
print(f"Mean Squared Error (average squared error): {mse:.2f}")
print(f"Root Mean Squared Error (average error magnitude): {rmse:.2f}")
```

# 5. Visualize Residuals

```python
plt.figure(figsize=(6, 4))
plt.scatter(X, residuals, color='green', label='Residuals')
plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.title('Residuals of the Regression Model')
plt.xlabel('X')
plt.ylabel('Residuals')
plt.legend()
plt.grid(True)
plt.show()
```
