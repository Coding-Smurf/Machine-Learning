# 1. Import Libraries and Generate Data

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_classification

# Generate synthetic regression data
X_raw, y = make_classification(n_samples=100, n_features=2, n_classes=2, n_informative=1, n_redundant=0, n_repeated=0, n_clusters_per_class=1, random_state=0, class_sep=0.3)
```

# 2. Calculate Logistic Regression Coefficients

```python
from sklearn.linear_model import LogisticRegression

# Fit a logistic regression model
model = LogisticRegression()
model.fit(X_raw, y)

# Get the coefficients and intercept
coefficients = model.coef_
intercept = model.intercept_

# Get the data points for plotting the decision boundary
x_vals = np.linspace(X_raw[:, 0].min(), X_raw[:, 0].max(), 100)
y_vals = -(intercept[0] + x_vals * coefficients[0][0]) / coefficients[0][1]
```

# 3. Visualize Data and Regression Line

```python
plt.figure(figsize=(6, 4))
plt.scatter(X_raw[:, 0], X_raw[:, 1], c=y, cmap='bwr')
plt.plot(x_vals, y_vals)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Decision Boundary")
plt.show()
```

# 4. Calculate Performance Metrics

```python
from sklearn.metrics import classification_report

# Predictions (Precision, Recall, F1-Score)
y_pred = model.predict(X_raw)
print(classification_report(y, y_pred))
```
