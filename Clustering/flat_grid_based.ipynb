# Grid Based Clustering (Flat/Partitional)

Instead of working directly on individual data points, grid-based methods work on the cells, which makes them typically very fast and scalable, especially for large datasets. Grid-based clustering is a type of clustering technique that uses a grid structure to partition the data space into a finite number of cells (or bins) and then performs clustering operations based on these cells.

### CLIQUE

QLIQUE is a grid-based clustering algorithm designed to quickly identify clusters in high-dimensional data by partitioning the data space into grids and analyzing dense cells.

```python
from pyclustering.cluster.clique import clique, clique_visualizer
from sklearn.datasets import make_moons

# Generate synthetic data
X, _ = make_moons(n_samples=300, noise=0.05, random_state=0)

# Create a CLIQUE instance 
clique_instance = clique(X.tolist(), 7, 1)
clique_instance.process()

# Retrieve results
clusters = clique_instance.get_clusters()
noise = clique_instance.get_noise()
cells = clique_instance.get_cells()

# Visualize the grid formed by CLIQUE
clique_visualizer.show_grid(cells, X.tolist())
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.array(X)
colors = ['blue', 'green']

plt.figure(figsize=(5,4))

for i, cluster in enumerate(clusters):
    cluster_points = data[cluster]
    plt.scatter(*cluster_points.T, color=colors[i % len(colors)])

if noise:
    noise_points = data[noise]
    plt.scatter(*noise_points.T, color='black', marker='x')

plt.title('CLIQUE Clustering Results')
plt.show()
```
